```python
import pandas as pd
df = pd.read_parquet('2024.parquet')
df
```

```python
df['return'] = df['close'].astype(float).diff(1).shift(-1) / df['close'].astype(float)
df
```

```python
import pandas as pd
import numpy as np \

def calculate_scores(window, df):
    df = df.copy()[:]
    reversed_data = df['return'][::-1]
    df['CRscore'] = (reversed_data+1).rolling(window).sum()[::-1]
    df['Conscore'] = np.sign(reversed_data).rolling(window).sum()[::-1]
    # print(df.iloc[:,-2:])
    return (df['CRscore']*df['Conscore'])/ window**2


calculate_scores(5, df).plot.kde()
# count = len([x for x in calculate_scores(10, df) if abs(x) > 0.5])
# print(count)
# for i in range(2, 1000):
#     print(calculate_scores(i, df).mean())
```
