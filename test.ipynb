```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt


def Sharpe(pnl:pd.Series)->float:
    sharpe = pnl.mean() / pnl.std()
    return sharpe

def calculate_Cr(df:pd.DataFrame, window:int)->pd.Series:
    df['return'] = df['close'].astype(float).diff(1).shift(-1) / df['close'].astype(float)
    reversed_data = df['return'][::-1]
    Cr = (reversed_data).rolling(window).sum()[::-1]
    return Cr

def generate_signal(df:pd.DataFrame)->pd.Series:
    N = 20
    k = 2
    Ma = df['close'].rolling(N).mean()
    Sd = df['close'].rolling(N).std()
    Mb = Ma + k*Sd
    Dn = Ma - k*Sd
    raw_Signal = (df['close'] < Dn) * 0.1
    
    # raw_Signal = -(df['close'].astype(float)/df['open'].astype(float) -1)*100
    # raw_Signal = raw_Signal *(raw_Signal>0)
    return raw_Signal

def position_restrict(Signal:pd.Series, window):
    Close_position = Signal.shift(window).fillna(0)
    Position = (Signal - Close_position).cumsum()
    Restrict = abs(Position) <= 1
    new_Signal = (Signal * Restrict).astype(float)
    return new_Signal
    

def calculate_pnl(Position:pd.Series, Cumulative_Return:pd.Series):
    Cumulative_Return = Cumulative_Return.dropna()
    Position = Position.reindex_like(Cumulative_Return)
    pnl = Position * Cumulative_Return
    return pnl
    
def calculate_cost(Position:pd.Series):
    cost = abs(Position.diff()) * 0.00075
    return cost

def main():
    df = pd.read_parquet('2024.parquet')[:5000]
    window = 20
    Cr = calculate_Cr(df, 5)
    row_Signal = generate_signal(df)
    Signal = position_restrict(row_Signal, window)
    pnl = calculate_pnl(Signal, Cr)
    pnl.index = pd.to_datetime(pnl.index, unit='ms')
    Position = (Signal - Signal.shift(window).fillna(0)).cumsum()
    Position.index = pd.to_datetime(Position.index, unit='ms')
    Position = Position.reindex_like(pnl)
    cost = calculate_cost(Position)
    pnl_net = pnl - cost
    
    plt.figure(figsize=[32, 16])
    p1 = plt.subplot(211)
    p2 = plt.subplot(212)
    p1.plot(pnl.fillna(0).cumsum(), label=f"PNL, Sharpe: {Sharpe(pnl)}")
    p1.plot(pnl_net.fillna(0).cumsum(), label=f"PNL_Net, Sharpe: {Sharpe(pnl_net)}")
    p1.grid(linestyle='--')
    p1.legend(loc='upper left')
    p2.plot(Position, label=f"PNL_Net")
    p2.grid(linestyle='--')
    p2.legend(loc='upper left')
    plt.show()
    
if __name__ == '__main__':
    main()
```

```python
import numpy as np 
import pandas as pd
np.random.seed(42)

def take_profitANDstop_loss(close, order_volume, up:float, low:float):
    mask = np.triu(np.ones((len(close),len(close)), dtype=bool))
    condition_up = np.where(mask, np.tile(close*up,[len(close), 1]).T, np.nan) #止盈矩阵，每一列都是行时刻下单对应止盈止损点，且是上三角阵，即只判断下单后的，下单前都为0
    condition_low = np.where(mask, np.tile(close*low,[len(close), 1]).T, np.nan) #止损矩阵，每一列都是行时刻下单对应止盈止损点，且是上三角阵，即只判断下单后的，下单前都为0
    exit_time_up = pd.DataFrame(condition_up <= np.array(close)).idxmax(axis=1).replace(0, np.inf)  #每个时刻close与所有时刻止盈点比较，找到第一次大于止损点时刻，得到Series
    exit_time_low = pd.DataFrame(condition_low >= np.array(close)).idxmax(axis=1).replace(0, np.inf) #每个时刻close与所有时刻止损点比较，找到第一次小于止损点时刻，得到Series
    exit_time = exit_time_up.combine(exit_time_low, min)#先止盈还是先止损
    pos_matrix = pd.DataFrame(np.tile(range(len(close)), [len(close),1])).T #位置矩阵，第n行为全n 行向量
    tile_pos = pd.DataFrame(np.tile(order_volume,[len(close),1])) #初始订单量拓展矩阵，每一列是列时刻订单量的常向量
    adjust = np.tril(tile_pos * (pos_matrix == exit_time)).sum(1) #通过广播将位置矩阵与平仓时刻比较，每一列是 列时刻订单在行时刻平仓数量的列向量, 将行时刻需要平仓数量加总，即可得到每时刻需平仓量
    order_volume = pd.DataFrame(order_volume - adjust)
    order_volume.index = close.index
    return order_volume


close = pd.read_parquet('BTCUSDT.parquet')['close'][:10000]
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close))
take_profitANDstop_loss(close, order_volume, up, low)
```

```python
import numpy as np
import pandas as pd
import time

np.random.seed(42)

def take_profitANDstop_loss_optimized(close, order_volume, up: float, low: float):
    close_array = close.to_numpy()
    len_close = len(close_array)
    
    # 计算止盈和止损点
    condition_up = close_array * up
    condition_low = close_array * low

    # 初始化平仓数量调整数组
    adjust = np.zeros(len_close)

    # 使用更高效的方法进行查找
    exit_times = np.full(len_close, np.inf)
    
    for i in range(len_close):
        up_condition_met = np.searchsorted(close_array[i:], condition_up[i], side='left') + i
        low_condition_met = np.searchsorted(close_array[i:], condition_low[i], side='right') + i

        if up_condition_met < len_close and close_array[up_condition_met] >= condition_up[i]:
            exit_time_up = up_condition_met
        else:
            exit_time_up = np.inf

        if low_condition_met < len_close and close_array[low_condition_met] <= condition_low[i]:
            exit_time_low = low_condition_met
        else:
            exit_time_low = np.inf

        # 选择止盈或止损中较早发生的时间
        exit_times[i] = min(exit_time_up, exit_time_low)

    # 累加平仓数量
    valid_exit_times = exit_times[exit_times < len_close].astype(int)
    np.add.at(adjust, valid_exit_times, order_volume[exit_times < len_close])

    # 计算调整后的订单量
    order_volume_adjusted = order_volume - adjust

    # 构建结果 DataFrame
    order_volume_df = pd.DataFrame(order_volume_adjusted, index=close.index)
    
    return order_volume_df

start_time = time.time()
close = pd.read_parquet('BTCUSDT.parquet')['close'][:100000]
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close))
take_profitANDstop_loss_optimized(close, order_volume, up, low)
end_time = time.time()
print("Execution time:", end_time - start_time)
```

```python
import numpy as np
import pandas as pd
import time

np.random.seed(42)

def take_profitANDstop_loss_optimized(close, order_volume, up: float, low: float):
    # 使用输入的 numpy array 而不是转换
    close_array = close if isinstance(close, np.ndarray) else close.to_numpy()
    len_close, num_assets = close_array.shape
    
    # 计算止盈和止损点
    condition_up = close_array * up
    condition_low = close_array * low

    # 初始化平仓数量调整数组
    adjust = np.zeros((len_close, num_assets))

    # 使用广播和 NumPy 向量化操作来找到止盈和止损的时间
    up_mask = close_array >= condition_up
    low_mask = close_array <= condition_low

    exit_times_up = np.argmax(up_mask, axis=0)
    exit_times_low = np.argmax(low_mask, axis=0)

    # 如果没有找到满足条件的位置，将其设置为 len_close
    exit_times_up[~up_mask.any(axis=0)] = len_close
    exit_times_low[~low_mask.any(axis=0)] = len_close

    # 找到止盈和止损中较早的时间
    exit_times = np.minimum(exit_times_up, exit_times_low)

    # 累加平仓数量，使用 NumPy 的广播来加速
    valid_exit_mask = exit_times < len_close
    for j in range(num_assets):
        if valid_exit_mask[j]:
            np.add.at(adjust[:, j], exit_times[j], order_volume[exit_times[j], j])

    # 计算调整后的订单量
    order_volume_adjusted = order_volume - adjust

    # 构建结果 DataFrame
    if isinstance(close, pd.DataFrame):
        order_volume_df = pd.DataFrame(order_volume_adjusted, index=close.index, columns=close.columns)
    else:
        order_volume_df = order_volume_adjusted
    
    return order_volume_df

start_time = time.time()
close = np.random.rand(400000,500)
up = 1.00001
low = 0.99999
order_volume = np.random.rand(len(close),500)
take_profitANDstop_loss_optimized(close, order_volume, up, low)
end_time = time.time()
print("Execution time:", end_time - start_time)
```
