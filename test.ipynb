```python
import pandas as pd
import numpy as np \

def calculate_scores(window, df):
    df = df.copy()[:]
    reversed_data = df['return'][::-1]
    df['CRscore'] = (reversed_data+1).rolling(window).sum()[::-1]
    df['Conscore'] = np.sign(reversed_data).rolling(window).sum()[::-1]
    # print(df.iloc[:,-2:])
    return (df['CRscore']*df['Conscore'])/ window**2


calculate_scores(5, df).plot.kde()
# count = len([x for x in calculate_scores(10, df) if abs(x) > 0.5])
# print(count)
# for i in range(2, 1000):
#     print(calculate_scores(i, df).mean())
```

```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

def calculate_Cr(df:pd.DataFrame, window:int)->pd.Series:
    df['return'] = df['close'].astype(float).diff(1).shift(-1) / df['close'].astype(float)
    reversed_data = df['return'][::-1]
    Cr = (reversed_data).rolling(window).sum()[::-1]
    return Cr

def generate_signal(df:pd.DataFrame)->pd.Series:
    raw_Signal = -(df['close'].astype(float)/df['open'].astype(float) -1)*10
    return raw_Signal

def position_management(Signal:pd.Series, window):
    Close_position = Signal.shift(window).fillna(0)
    Position = (Signal - Close_position).cumsum()
    
    return Signal
    

def calculate_pnl(Position:pd.DataFrame, Cumulative_Return:pd.DataFrame):
    Cumulative_Return = Cumulative_Return.dropna()
    Position = Position.reindex_like(Cumulative_Return)
    pnl = Position * Cumulative_Return
    return pnl    

def main():
    df = pd.read_parquet('2024.parquet')
    window = 5
    Cr = calculate_Cr(df, 5)
    row_Signal = generate_signal(df)
    Signal = position_management(row_Signal, window)

    Position = (Signal - Signal.shift(window).fillna(0)).cumsum()
    Position.index = pd.to_datetime(Position.index, unit='ms')
    pnl = calculate_pnl(Signal, Cr)
    pnl.index = pd.to_datetime(pnl.index, unit='ms')
    plt.figure(figsize=[32, 16])
    p1 = plt.subplot(211)
    p2 = plt.subplot(212)
    p1.plot(pnl.fillna(0).cumsum(), label=f"PNL_Net")
    p1.grid(linestyle='--')
    p1.legend(loc='upper left')

    p2.plot(Position, label=f"PNL_Net")
    p2.grid(linestyle='--')
    p2.legend(loc='upper left')


    plt.show()
    
if __name__ == '__main__':
    main()
```
